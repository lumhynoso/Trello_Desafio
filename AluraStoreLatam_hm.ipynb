### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

# 2. Ventas por categoría

```python
#En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

def analyze_sales(df, store_name):
    # Productos más y menos vendidos
    product_sales = df['Categoría del Producto'].value_counts()
    top_product = product_sales.index[0]
    print(f"Tienda {store_name}:")
    print(f"Producto más vendido: {top_product}")
    return product_sales


sales_tienda1 = analyze_sales(tienda, 1)
sales_tienda2 = analyze_sales(tienda2, 2)
sales_tienda3 = analyze_sales(tienda3, 3)
sales_tienda4 = analyze_sales(tienda4, 4)



```

# 3. Calificación promedio de la tienda


```python

Pro_tienda =  tienda['Calificación'].mean()
Pro_tienda2 = tienda2['Calificación'].mean()
Pro_tienda3 = tienda3['Calificación'].mean()
Pro_tienda4 = tienda4['Calificación'].mean()

print(f"El promedio de la tienda 1 es: {Pro_tienda}")
print(f"El promedio de la tienda 2 es:{Pro_tienda2}")
print(f"El promedio de la tienda 3 es:{Pro_tienda3}")
print(f"El promedio de la tienda 4 es:{Pro_tienda4}")
```

```python

import matplotlib.pyplot as plt

promedios = [Pro_tienda, Pro_tienda2, Pro_tienda3, Pro_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 8))
plt.pie(promedios, labels=tiendas, autopct='%1.1f%%', startangle=150, colors=['skyblue', 'lightcoral', 'lightgreen', '#05FF80'])
plt.title('Calificación promedio de las tiendas')
plt.axis('equal')
plt.show()
```

# 4. Productos más y menos vendidos

```python
def analyze_sales(df, store_name):
    # Productos más y menos vendidos
    product_sales = df['Producto'].value_counts()
    top_product = product_sales.index[0]
    bottom_product = product_sales.index[-1]
    print(f"Tienda {store_name}:")
    print(f"Producto más Vendido +: {top_product} (Ventas: {product_sales[top_product]})")
    print(f"Producto menos Vendido -: {bottom_product} (Ventas: {product_sales[bottom_product]})")
    return product_sales


sales_tienda1 = analyze_sales(tienda, 1)
sales_tienda2 = analyze_sales(tienda2, 2)
sales_tienda3 = analyze_sales(tienda3, 3)
sales_tienda4 = analyze_sales(tienda4, 4)



```
